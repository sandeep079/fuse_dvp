# **What are Foundation Models?**

A foundation model refers to any model trained on diverse data through self-supervision at scale, allowing for adaptation, such as fine-tuning, across various downstream tasks. Examples include BERT, GPT-3, and CLIP. While not a new concept technologically, the recent surge in scale and breadth of these models, like GPT-3 with its 175 billion parameters, has expanded our perception of what is achievable. Despite their potential benefits, existing foundation models raise concerns about potential harms, and their characteristics are not fully understood, prompting intense scrutiny due to their imminent widespread deployment.

# **Emergence and Homogenization**

**Emergence** in AI refers to the phenomenon where the behavior of an artificial intelligence system is not explicitly programmed but emerges from the complexity of its components. For example, in a neural network, specific features or patterns may emerge during training that were not explicitly defined by the programmer. This can lead to the AI system exhibiting unexpected behaviors or insights that were not directly programmed into it.

**Homogenization** in AI involves the consolidation of methodologies for building machine learning systems across various applications. For example, OpenAI's GPT-3 (Generative Pre-trained Transformer 3) showcases homogenization as it can be adapted to a wide range of tasks through fine-tuning or prompt-based instructions. Despite its vast applicability, relying on a singular model for diverse tasks introduces a potential single point of failure. If there are vulnerabilities or biases in GPT-3, they could impact a broad spectrum of AI applications that leverage this homogenized approach.

# **Ecosystem of Foundation Models**
To really understand foundation models and how they're used, we need to look at the whole system they're a part offrom creating data to putting the model into action. Remember, the foundation model is a big deal, but it's just one piece of the whole AI setup. If we simplify things, we can break down the process into stages, like training and adapting the model. What's interesting is that people are involved at both ends of this process, and this way of looking at things helps us realize that different questions about foundation modelslike whether they're ethicalneed different answers at different stages.

1. **Data Creation:** This stage involves the human-centric process of creating data, which can be explicitly about people or measure aspects of the environment. All data has an owner and a purpose, which may or may not include training a foundation model.

2. **Data Curation:** Data is curated into datasets, acknowledging that there is no single natural distribution of data. Ensuring data relevance, quality, and compliance with legal and ethical constraints is crucial, yet often underappreciated in AI research.

3. **Training:** The celebrated centerpiece of AI research, training foundation models occurs on curated datasets. This stage involves teaching the model patterns and features from the data.

4. **Adaptation:** In machine learning research, adaptation entails creating a new model based on the foundation model for a specific task. For deployment, it involves building a system with various modules, rules, classifiers, and integration with complementary signals. This stage is crucial for mitigating potential harms, especially for models generating content that may be problematic.

5. **Deployment:** The direct social impact of an AI system is realized when it is deployed to people. While caution is necessary to avoid deploying potentially harmful models trained on questionable data, gradual releases in large-scale deployments can help mitigate potential harms. This stage marks the practical application of the AI system in the real world.
