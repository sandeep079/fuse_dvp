# **Background for Large Vision Models**

Since the introduction of the Transformer architecture by Vaswani et al., deep learning models have undergone significant advancements in both parameter size and complexity. Notable language models such as BERT, T5, GPT-1, and GPT-2, along with various domain-specific variants, have contributed to this progress. Large language models, like GPT-3, ChatGPT, and GPT-4, trained on extensive datasets through self-supervised learning, exhibit enhanced natural language comprehension and adaptability to diverse downstream tasks.

Inspired by the success of pre-trained language models in natural language processing, researchers have extended the concept to computer vision with pre-trained visual models like ViT, Swin Transformer, and VideoMAE V2. These models, pre-trained on massive image datasets, enable computers to understand image content and extract rich semantic information for various downstream applications. Multi-modal visual models, such as CLIP and ALIGN, use contrastive learning to align textual and visual information, facilitating efficient generalization in downstream tasks. Despite their achievements, these models still face limitations in terms of generalization capabilities.


The rapid progress in artificial intelligence (AI) has led to the development of AI systems based on foundational models, representing a significant area of research. This conceptual framework, coined by AI experts, emphasizes emergence in machine learning, where system behavior evolves dynamically through tasks like automatic inference and the progressive development of advanced features. Recently, the Segment Anything Model (SAM) introduced a trend in solving downstream tasks through prompt engineering modules, showcasing remarkable zero-shot generalization capabilities. Visual prompt engineering has become crucial for guiding large visual models in specific tasks, involving the design and optimization of visual prompts.

The emergence of foundation models, particularly in computer vision, has unlocked substantial potential. Visual prompt engineering serves as an adaptive interface, seamlessly integrating with large visual models. The synergy between large visual models and visual prompt engineering empowers the utilization of foundation models, enhancing flexibility and efficiency in image analysis and task resolution. This integration opens avenues for exploration in AI applications, presenting numerous prospects and unprecedented opportunities.

# **UNIFIED-IO**

The **UNIFIED-IO** model is designed to address a diverse set of tasks encompassing language, vision, and multi-modal challenges in a unified manner. To comprehensively evaluate its capabilities, the model undergoes multi-task training on **95 datasets** sourced from **62 publicly available data repositories**. These datasets span a wide spectrum of tasks, skills, and modalities.

## Task Categorization

The modalities of input and output for each task are categorized into four types:

- **Text:** Natural language tokens
- **Image:** RGB images
- **Sparse:** A small number of location coordinates within the image
- **Dense:** Per-pixel labels such as depth maps and surface normal maps

For ease of training and analysis, the datasets are grouped into eight categories and 22 tasks.

## Task Groups and Descriptions

1. **Image Synthesis:**
   - Generate a matching image based on inputs like a text description, partially occluded image with an inpainting target, or a segmentation map with semantic classes.
   - Data sources: Image and text pairs, bounding boxes, semantic segmentation.

2. **Sparse Labelling:**
   - Identify target regions or keypoint locations in an image based on a natural language query.
   - Tasks: Object detection, object localization, human pose estimation, referring expression.

3. **Dense Labelling:**
   - Produce per-pixel labels for an image (e.g., distance to camera, surface orientation, semantic class).

4. **Image Classification:**
   - Generate a class name or tag for an image or target region, optionally with a bounding box.
   - Includes image classification and object


# **Modalities Representation in UNIFIED-IO**

To support a variety of modalities such as images, language, boxes, binary masks, and segmentation masks, the **UNIFIED-IO** model employs a strategy of representing these modalities in a shared and unified space. This involves discretizing text, images, and other structured outputs, representing them using tokens drawn from a unified and finite vocabulary.

## Text Representation

Text inputs and outputs are tokenized using SentencePiece . Additionally, each task is specified with a natural language prompt (excluding some tasks like VQA) to indicate the intended operation. For instance, prompts like "What is the depth map of the image?" for depth estimation or "What region does 'cat' describe?" for object localization are utilized.

## Images and Dense Structures Representation

Tasks in computer vision requiring high-dimensional outputs, such as images and per-pixel labels, are addressed by converting per-pixel labels into RGB images. Depth maps are normalized to grayscale images, surface normal orientations are converted to RGB values, and segmentation instances are mapped to unique colors. The color-to-class mapping is specified in the text to avoid a fixed list of classes. Image tokens are then encoded using a VQ-GAN during training, with the VQ-GAN codebook added to the vocabulary. During inference, the VQ-GAN decoder converts generated image tokens into the final output image.

## Sparse Structures Representation

Sparse structures like bounding boxes or human joints are encoded using special tokens added to the vocabulary to represent discretized image coordinates. Bounding boxes are encoded using sequences of four tokens, and labeled boxes include a text class label. Human joints are encoded as sequences of points with associated text visibility labels. This approach enables handling a wide range of tasks utilizing these elements in their inputs or outputs.



# **Example Of Large Vision Models**

Let's look at various different Vision Models available currently


## **Transformer**

The Transformer architecture has become a foundational model, revolutionizing the field with its initial application in translation tasks in natural language processing (NLP). Starting with Transformer-based BERT, its versatility and scalability led to the development of larger models such as GPT-1, GPT-2, GPT-3, GPT-4, T5, PaLM, LLaMA, and others, significantly advancing performance and generalization capabilities, even surpassing human-level performance in certain tasks. The Vision Transformer (ViT) extended Transformer's application to computer vision, successfully bridging the gap between textual and image domains.

In the Transformer architecture, the smallest unit is a token, making it well-suited for handling multi-modal data by converting different modalities into tokens. This characteristic has been leveraged in various multi-modal works like ViTL, DALL-E, CLIP, VLMO, ALBEF, and others, which employ Transformer as the primary framework for tasks like text-to-image and image-to-text retrieval, image captioning, and image/text generation. With the advent of large models, researchers have proposed multi-modal models like CoCa, Flamingo, BEiTv3, PALI, and GPT-4, aiming to further enhance performance across various downstream tasks. In summary, the Transformer remains a dominant and foundational model in contemporary research across different domains.

## **CLIP**

OpenAI has introduced a revolutionary vision-language model called CLIP (Contrastive Language-Image Pre-Training), leveraging the correlation between images and text for weakly supervised pre-training. This approach significantly enhances performance by expanding the available data, with a massive dataset of 400 million image-text pairs collected for training. CLIP utilizes fixed human-designed prompts, enabling zero-shot prediction and showcasing superior few-shot capabilities compared to other state-of-the-art models. Its success emphasizes the effectiveness of combining visual and textual information and the power of weakly supervised training with large datasets. CLIP's achievements signal a potential breakthrough in multi-modal techniques, offering insights into capturing rich feature representations. Additionally, it inspires advancements in prompt engineering in computer vision, opening promising avenues for future developments. Other models like Group VIT, ViLD, Glip, Clipasso, Clip4clip, and ActionCLIP further contribute to this evolving landscape.

## **VPT**


In the adaptation of large vision models to downstream tasks, a preferred approach is to modify the input rather than altering the pre-trained model's parameters. This method involves introducing a small number of task-specific learnable parameters into the input space, enabling the learning of task-specific continuous vectors. Known as prompt tuning, this technique allows for efficient fine-tuning of pre-trained models without modifying their underlying parameters. The VPT (Visual Prompt Tuning) approach was the first to explore the universality and feasibility of visual prompts. It includes both deep and shallow versions, achieving impressive results by learning the prompt and class head of the input data while keeping the pre-trained transformer model parameters fixed. VPT aimed to showcase the effectiveness of visual prompting and introduced a novel perspective on prompt design. By demonstrating satisfactory results through input modification alone, VPT highlighted the potential of prompt tuning as an efficient strategy for downstream tasks, paving the way for further research in prompt engineering.

## **SAM**

In 2023, Meta AI introduced SAM, a project focused on creating a universal image segmentation model capable of addressing diverse downstream tasks through prompt engineering. SAM utilizes prompt segmentation as a pre-training objective to enhance its flexibility and robustness against interference. The model is divided into three components: image encoder, prompt encoder, and mask decoder, efficiently distributing computational cost for adaptability and versatility. SAM's strength lies in its efficient generalization across various segmentation tasks, achieved through prompt engineering. The methodology involves pre-training on prompt segmentation and fine-tuning on specific downstream tasks, allowing SAM to leverage learned knowledge for improved performance in tasks such as medical image analysis, video object tracking, data annotation, 3D reconstruction, robotics, and image editing. SAM's modular design enables flexibility and adaptability to different prompt formats, making it a versatile solution for diverse segmentation challenges.
