# **Increase of Model Parameters Over The Years**

Over the years, there has been a noticeable trend in the field of large language models (LLMs) where the number of parameters in these models has experienced a substantial increase. The term "parameters" refers to the internal variables that the model adjusts during training to make predictions and generate responses. In the context of LLMs, these parameters are the weights associated with the connections between neurons in the neural network.

The growth in the number of model parameters is indicative of the increasing complexity and capacity of these language models. Larger models have more parameters, allowing them to capture and represent more intricate patterns, relationships, and nuances in the data they are trained on. This results in enhanced capabilities in tasks such as language understanding, generation, translation, summarization, and more.

The advantages of this growth in model parameters can be summarized as follows:

1. Improved Representation of Knowledge: More parameters enable the model to learn and encode a richer representation of the complexities present in natural language. This means the model can understand and generate more nuanced and contextually relevant outputs.

2. Enhanced Generalization: Larger models often exhibit better generalization to diverse and complex datasets. They can adapt to a wider range of inputs and perform well across a variety of tasks.

3. Increased Learning Capacity: The higher number of parameters allows the model to store and remember a larger amount of information during training, which can be beneficial for tasks that require a deep understanding of context or long-range dependencies.

4. Fine-tuning for Specific Tasks: The increased capacity allows for effective fine-tuning on specific tasks. Pre-training on a large corpus of data followed by fine-tuning on a narrower dataset for a specific task has become a common practice, leading to improved performance.

# **Increase of Context Length Over The Years**

 The evolution of model context length in recent years has been a crucial aspect of the advancements in large language models. If we examine the trajectory of various models, there's a noticeable and substantial increase in the length of context they can handle.

For instance, GPT-3, a notable predecessor, boasted an impressive context length of 2 thousand tokens. This meant that the model could consider and analyze information within a 2 thousand-token window, which is already substantial compared to earlier models. However, the recent release of GPT-4 Turbo has pushed the boundaries further, offering an extraordinary context length of 32 thousand tokens.

This increase in context length signifies the capacity of these models to comprehend and process much larger pieces of information in a single pass. To put it into perspective, GPT-4 Turbo can theoretically ingest the entirety of a 300-page novel at once, enabling a more comprehensive understanding of the context and relationships within a more extensive body of text.

The benefits of this expanded context length include:

1. Improved Long-Term Dependencies: Models can capture and understand longer sequences of information, allowing them to establish more robust connections and dependencies over extended passages of text.

2. Enhanced Contextual Understanding: The ability to consider a larger context enables models to grasp the broader meaning and nuances present in text, making them more adept at tasks that require a deep understanding of context.

3. Efficient Processing of Lengthy Documents: With the capability to handle extensive context, these models become more efficient at processing and generating responses for longer documents or conversations.

# **Disadvantages of Model Scaling**

It's essential to consider that increasing context length also poses challenges, particularly in terms of computational requirements and memory constraints. Handling such large context lengths demands significant computational resources during both training and inference.

The significant increase in model context length, as exemplified by GPT-4 Turbo, reflects a remarkable stride in the capability of large language models to process and comprehend extensive textual information, opening up new possibilities for applications in natural language understanding, summarization, and other language-related tasks.

# **Monarch Mixture Model**

Monarch Mixer (M2), a novel machine learning architecture that scales sub-quadratically along both sequence length and model dimension. M2 utilizes Monarch matrices, a class of expressive structured matrices, achieving high hardware efficiency on GPUs. In experiments across non-causal BERT-style language modeling, ViT-style image classification, and causal GPT-style language modeling, M2 performs competitively with existing architectures while using fewer parameters. For instance, in non-causal BERT-style modeling, M2 matches the quality of BERT-base and BERT-large with up to 27% fewer parameters and higher throughput. In image classification on ImageNet, M2 outperforms ViT-b by 1% with only half the parameters. In causal GPT-style models, M2 overcomes the quadratic bottleneck introduced by causality, showcasing promising results in matching Transformer quality without attention or MLPs at 360M parameters in pretraining perplexity on The PILE.

##**Monarch Matrix**
A Monarch matrix is a type of structured matrix that can represent many linear transforms, such as attention, convolution, and Fourier transform. It is defined as a product of two block-diagonal matrices, which makes it hardware-efficient and scalable. Monarch matrices can be used to train and fine-tune large neural networks in various domains, such as natural language processing, computer vision, and physics.

# M2-BERT( A application of Monarch Matrix)
Untitled.jpg

*Fu, D., Arora, S., & R, C. (2023). Monarch Mixer: Revisiting BERT, without attention or MLPs. Hazy Research*

M2-BERT serves as a direct alternative to BERT-style language models. The training of M2-BERT involves masked language modeling on the C4 dataset, utilizing the bert-base-uncased tokenizer. The fundamental concept is to take a conventional Transformer structure and replace attention mechanisms and MLPs with M2 layers. In the sequence mixer, bidirectional gated convolutions and a residual convolution replace attention, employing Monarch matrices set to DFT and inverse DFT forms. Additionally, short depthwise convolutions are introduced after projections. In the dimension mixer, dense matrices within MLPs are substituted with learned block-diagonal matrices. M2-BERT models, available in both base and large versions, undergo pretraining with sizes corresponding to the standard BERT-base and BERT-large models.

- M2-BERT uses Monarchs to perform FFTs for long convolutions, replacing traditional FFT computations.

- M2-BERT makes convolutions non-causal by doubling the length of the convolution kernel, allowing weights to wrap around.

- An extra convolution (a "residual") is added in M2-BERT, enhancing performance on synthetic tasks and pretraining loss.

- During fine-tuning, M2-BERT improves on BERT's approach by using average pooling of embeddings in the input, particularly beneficial for tasks involving information comparison across multiple sentences, such as GLUE NLI tasks.

# **Results of Monarch Mixture Architecture**


The Monarch Mixer utilizes Monarch matrices and elementwise operations to create architectures with sub-quadratic complexity. This method views each layer as a series of mixing operations across both sequence and model dimension axes. The input for each layer is a sequence of embeddings, and the output is another sequence, with parameters defining sequence length and model dimension.

By employing order-2 Monarch matrices, the Mixer is capable of constructing expressive architectures, such as convolutional blocks featuring sparse MLPs. These architectures are created through mixing operations along both the sequence and embedding axes. For instance, specifying the first matrix as the Discrete Fourier Transform and the second matrix as its inverse corresponds to convolutions in frequency space. Furthermore, the Mixer can emulate an MLP, with dense matrices being replaced by Monarch matrices.
